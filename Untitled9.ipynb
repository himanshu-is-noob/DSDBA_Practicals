```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
df = pd.DataFrame(data = titanic )
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop('deck' , axis = 1)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['age'].fillna(df['age']).mean()
```

```python
# People who survived based on their gender
```

```python
sns.scatterplot(data = df , x = 'sex' , y = 'age' , hue = 'survived')
```

```python
sns.countplot(data =df , x = 'age' )
```

```python
sns.histplot(data = df , x = 'class' , y = 'fare' , hue = 'survived'  )
```

```python
sns.kdeplot(data = df , x = 'age' , hue = 'survived' , multiple = 'stack')
```

```python
sns.displot(data = df , x = 'fare' , hue = 'survived' ,bins = 2, kde = True)
```

```python
# Now we need to check how ticket fare is distributed
```

```python
sns.histplot(data = df , x = 'fare' , kde= True , bins = 20  )
```

```python
sns.histplot(data = df , x = 'fare' , kde= True , bins = 20 , hue = 'sex' )
```

```python
sns.histplot(data = df , x = 'class' , kde = True)
plt.title("Classwise Distributon ")
plt.xlabel("Class")
plt.ylabel("No of Passenger")
```

```python
sns.histplot(data = df , x = 'fare' , hue = 'survived' , bins =50 , kde = True)
plt.title('No of People vs Fare based on Survival')
plt.xlabel('Ticket Prices')
plt.ylabel('No of People')
```
