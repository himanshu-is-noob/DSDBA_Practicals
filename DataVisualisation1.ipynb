```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df = df[['survived' , 'class' , 'sex' , 'age' , 'fare' , 'alive' , 'embark_town', 'pclass']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.jointplot(x='age' , y = 'fare' , data = df , hue= 'survived')
```

```python
sns.boxplot(df['age'])
```

```python
sns.histplot(data = df , x = "sex" )
```

```python
sns.barplot(x = 'sex' , y = 'survived' , hue='class' , data = df) 
```

```python
sns.kdeplot(df['fare'])
```

```python
sns.countplot(x = df['class'])
```

```python
sns.countplot(data = df , x = 'survived', palette = 'pastel' )
```

```python
sns.histplot(data = df , x = 'fare' , kde = True , bins = 30 , color = 'blue')
plt.title('Distribution of Titanic Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
```

```python
#Survival by Gender 

sns.countplot(data = df , x = 'sex' , hue = 'survived' , palette = 'pastel')
```

```python
# Passenger  Class 

plt.figure(figsize = (10,6))
sns.countplot(data = df , x = 'pclass' , hue = 'survived' , palette = 'Set1' )
plt.title("Survival Count by Survival Class")
plt.xlabel('Passenger Class')
plt.ylabel('Count of People')
plt.show()
```

```python
# age vs fare 

plt.figure(figsize= (10,6))

sns.scatterplot(data = df , x = 'age' , y ='fare' , hue= 'survived' )
plt.xlabel('age')
plt.ylabel('fare')
plt.title("Age vs Fare")
plt.show()
```

```python
sns.histplot(data = df , x = "fare" , kde = True , color ='blue' , bins = 40)
plt.xlabel('age')
plt.ylabel('count as per ticket prices')
plt.show()
```
