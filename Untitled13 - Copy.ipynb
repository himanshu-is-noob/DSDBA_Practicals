```python
# Now we need to fill the Null Values with mean values 
```

```python
df['Attendance (%)'] = df['Attendance (%)'].fillna(df['Attendance (%)'].mean())
```

```python
df['Assignments_Avg'] = df['Assignments_Avg'].fillna(df['Assignments_Avg'].mean())
```
